```python
import pymysql

db=pymysql.connect(host='localhost',user='root',passwd='0831',db='FilmRateAI',charset='utf8')
cursor=db.cursor()
```

```python
# 모든 결과를 한 번에 가져오기
sql = cursor.execute("select * from Reviews")
result = cursor.fetchall()
db.commit()
db.close()
```

```python
import pandas as pd
columns=['id','movie_id','review','star','positive_probability','new_star']
df=pd.DataFrame(result,columns=columns)
df.info()
```

```python
df.head()
```

```python
reviews0to2 = df.loc[(df["new_star"] >= 0) & (df["new_star"] <=2), "review"].tolist()
type(reviews0to2)
reviews0to2
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import pandas as pd


model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "사용자 말투를 살려서 입력하는 리뷰 데이터를 요약하여 5줄로 출력하세요"),
        ("human", "{reviews}"),
    ]
)

chain = prompt | model | StrOutputParser()
```

```python
chain.invoke({"reviews": reviews0to2})
```

```python
reviews3to4 = df.loc[(df["new_star"] >= 3) & (df["new_star"] <=4), "review"].tolist()
type(reviews3to4)
reviews3to4
```

```python
chain.invoke({"reviews": reviews3to4})
```

```python
reviews9to10 = df.loc[(df["new_star"] >= 9) & (df["new_star"] <=10), "review"].tolist()
type(reviews9to10)
reviews9to10
```

```python
summary10=chain.invoke({"reviews": reviews9to10})
```
